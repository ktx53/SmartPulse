```python
import requests
import json
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns

# API'ye GET isteği gönderme ve başlama bitiş parametreleri belirlenmesi
url = 'https://seffaflik.epias.com.tr/transparency/service/market/intra-day-trade-history'
start_date = '2022-01-26'
end_date = '2022-01-26'
params = {'startDate': start_date, 'endDate': end_date}
response = requests.get(url, params=params)

# API'den gelen verilerinin json formatını çevrilmesi
data = json.loads(response.text)

# Conract değeri PH ile başlayan sınıfların listesi
ph_conracts = [d for d in data['body']['intraDayTradeHistoryList'] if d['conract'].startswith('PH')]

# Her bir conract değeri için istenilen değerlerin hesaplanması

results = []
for conract in ph_conracts:
    # Tarih değerini alma
    Tarih = datetime.strptime(conract['conract'][2:], '%y%m%d%H')
    
    # Toplam işlem tutarı hesaplama
    Toplam_İslem_tutari = sum([d['price'] * d['quantity'] / 10 for d in ph_conracts if d['conract'] == conract['conract']])
    
    # Toplam işlem miktarı hesaplama
    Toplam_İslem_Miktari = sum([d['quantity'] / 10 for d in ph_conracts if d['conract'] == conract['conract']])
    
    # Ağırlıklı ortalama fiyat hesaplama
    Agirlikli_ort_Fiyat = Toplam_İslem_tutari / Toplam_İslem_Miktari
    
    # Sonuçları listeye ekleme
    results.append({'Conract': conract['conract'], 'Tarih': Tarih, 'Toplam İşlem Tutarı': Toplam_İslem_tutari, 'Toplam İşlem Miktarı': Toplam_İslem_Miktari, 'Ağırlıklı Ortalama Fiyat': Agirlikli_ort_Fiyat})
    
```

```python
#Hesaplanan conract değerlerini dataframe çevirip listelenmesi
results_df = pd.DataFrame(results, columns=['Tarih', 'Toplam İşlem Tutarı', 'Toplam İşlem Miktarı', 'Ağırlıklı Ortalama Fiyat'])
print(results_df.to_string(index=False))
```

```python
#Hesaplanan conract değerlerinin aynı olanlarının silinip dataframe çevirip listelenmesi

unique_results_df = results_df.drop_duplicates(subset=['Tarih', 'Toplam İşlem Tutarı', 'Toplam İşlem Miktarı', 'Ağırlıklı Ortalama Fiyat'])
print(unique_results_df.to_string(index=False))
```

```python
#Hesaplanan ortalama fiyatların görselleştirilmesi
plt.figure(figsize=(14,7))
sns.lineplot(data=unique_results_df, x="Tarih", y="Ağırlıklı Ortalama Fiyat")
plt.title("Tarih'e Göre Ağırlıklı Ortalama Fiyatların Değişimi")
plt.show()
```
